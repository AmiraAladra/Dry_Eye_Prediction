```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Define selected features for Sleep Quality and Dry Eye Disease
sleep_features = ['Wake up during night', 'Feel sleepy during day', 'Smoking',
       'Ongoing medication', 'Smart device before bed', 'Blue-light filter',
       'Dry Eye Disease', 'Gender_M', 'Stress level_3', 'Stress level_5']
dry_eye_features = ['Smart device before bed', 'Discomfort Eye-strain', 'Redness in eye',
       'Itchiness/Irritation in eye', 'Gender_F', 'Gender_M', 'Stress level_1',
       'Stress level_2', 'Stress level_3', 'Stress level_5']
```

```python
# Define target variables
sleep_target = 'Sleep quality'
dry_eye_target = 'Dry Eye Disease'
```

```python
# Split data for Sleep Quality Prediction
x_train_sq, x_test_sq, y_train_sq, y_test_sq = train_test_split(
    df[sleep_features], df[sleep_target], test_size=0.2, random_state=42, stratify=df[sleep_target]
)

# Split data for Dry Eye Disease Prediction
x_train_de, x_test_de, y_train_de, y_test_de = train_test_split(
    df[dry_eye_features], df[dry_eye_target], test_size=0.2, random_state=42, stratify=df[dry_eye_target]
)
```

```python
# Standardize features for Sleep Quality model
scaler_sq = StandardScaler()
x_train_sq_scaled = scaler_sq.fit_transform(x_train_sq)
x_test_sq_scaled = scaler_sq.transform(x_test_sq)

# Standardize features for Dry Eye model
scaler_de = StandardScaler()
x_train_de_scaled = scaler_de.fit_transform(x_train_de)
x_test_de_scaled = scaler_de.transform(x_test_de)
```

```python
# Logistic Regression for Sleep Quality
lr_sq = LogisticRegression()
lr_sq.fit(x_train_sq_scaled, y_train_sq)
y_pred_sq = lr_sq.predict(x_test_sq_scaled)
print("Logistic Regression - Sleep Quality Accuracy:", accuracy_score(y_test_sq, y_pred_sq))
```

```python
# Random Forest for Sleep Quality
rf_sq = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
rf_sq.fit(x_train_sq, y_train_sq)
y_pred_rf_sq = rf_sq.predict(x_test_sq)
print("Random Forest - Sleep Quality Accuracy:", accuracy_score(y_test_sq, y_pred_rf_sq))
```

```python
# Logistic Regression for Dry Eye
lr_de = LogisticRegression()
lr_de.fit(x_train_de_scaled, y_train_de)
y_pred_de = lr_de.predict(x_test_de_scaled)
print("Logistic Regression - Dry Eye Accuracy:", accuracy_score(y_test_de, y_pred_de))
print(confusion_matrix(y_test_de, y_pred_de))
```

```python
# Random Forest for Dry Eye
rf_de = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
rf_de.fit(x_train_de, y_train_de)
y_pred_rf_de = rf_de.predict(x_test_de)
print("Random Forest - Dry Eye Accuracy:", accuracy_score(y_test_de, y_pred_rf_de))
print(confusion_matrix(y_test_de, y_pred_rf_de))
```
